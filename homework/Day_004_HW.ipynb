````
#answer
import numpy as np

a = np.array( [20,30,40,50] )
b = np.array( [1,2,3,4] ) 
c = 1
d = np.array( [1] )
e = np.array( [1,2] ) 

print(a + a)
print(a + b)
print(a + c)
print(a + d)
print(a + e)
````

2. 如何在不用迴圈的情況下計算 (A+B)*(-A/2) ？那用迴圈怎麼做？

````
#answer
A = np.ones(3)*1
B = np.ones(3)*2

print((A + B) * (-A / 2))
````

3. 請問如何計算「1x6 的單位矩陣」和「6x1 的單位矩陣」的內積和外積？

````
# answer

A = np.ones((1,6))
B = np.ones((6,1))

print(A * B)
print(A @ B)
````
