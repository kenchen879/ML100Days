### 1. 請比較對一個 100 x 100 x 100 的陣列，使用不同方法對每一個元素 +1 的時間比較。

```python
import numpy as np
```

```python
Z = np.random.randint(0, 100, 1000000).reshape(100, 100, 100)
```

```python
%timeit -n 10 a = 2
for i in Z:
    for j in i:
        for k in j:
            i = i + 1
```

```python
%timeit -n 10 a = 2
for i in Z.flat:
    i = i+1
```

```python
%timeit -n 10 a = 2
for i in np.nditer(Z):
    i = i+1
```

### 3. 如何利用 list(...) 實現 a.tolist() 的效果？試著用程式實作。

```python
a = np.random.randint(10, size=6) 

print(a.tolist())
print(list(a))


b = np.random.randint(10, size=(3,4)) 

print(b.tolist())
print(list(b))


c = np.random.randint(10, size=(2,3,2)) 

print(c.tolist())
print(list(c))
```

```python
## 進階版

def tolist(iterable):
    if type(iterable) != np.ndarray:
        return iterable
    return [tolist(obj) for obj in iterable]

print(tolist(a))
print(tolist(b))
print(tolist(c))
```
